```python
from abc import ABC , abstractmethod
```

```python
class Runnable (ABC):
 @abstractmethod
 def invoke (input_data):
  pass
```

```python
import random

class nakliLLM(Runnable):
  def __init__(self):
    print ("LLM created")

  def invoke (self, prompt):

    response_list = [
        "adnan is gret",
        "SK is good",
        "ALI is "
    ]

    return {'response' : random.choice(response_list)}

  def predict (self, prompt):
    response_list = [
        "adnan is gret",
        "SK is good",
        "ALI is "
    ]

    return {'response' : random.choice(response_list)}



```

```python
llm = nakliLLM()
```

```python
llm.predict("what is your name")
```

```python
class nakliPrompt(Runnable):
  def __init__(self, input_variables, template):
    self.input_variables = input_variables
    self.template = template

  def invoke (self, input_dict):
    return self.template.format(**input_dict)
```

```python
class nakliLLMchain :
  def __init__(self, llm, prompt):
    self.llm = llm
    self.prompt = prompt

  def run(self, input_dict):
    prompt = self.prompt.format(input_dict)
    result = self.llm.predict(prompt)
    return result['response']
```

```python
class RunnableConnector (Runnable):
  def __init__(self, runnable_list):
    self.runnable_list = runnable_list
  def invoke (self, input_data):
    for runnable in self.runnable_list:
      input_data = runnable.invoke(input_data)
      return input_data
```

```python
class NakliStroutput(Runnable):
  def __init__(self):
    pass
  def invoke(self, input_data):
    return input_data['response']
```

```python
template = nakliPrompt(
    template = "what a poem about {topic}",
    input_variables = ['topic']
)

```

```python
parser = NakliStroutput()
```

```python
llm = nakliLLM()
```

```python
chain = RunnableConnector([template, llm, parser])
chain.invoke ({'topic': 'Pakistan'})
```
